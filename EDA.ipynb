Next step is to remove OUTLIERS

Five number summary 
1. minimum
2. maximum
3. media 
4. third quartile(Q3)
5. first quartile (Q1)

Quartile 1 is 25% of the data 
third quartile is 75% of the data 
Median is the centre value or point of the data ---> is used when discontinuous value of the data are present (discontinuous interval)
mean is used when you have continuous value of the data (continuous interval)

Formula 
Q1= 25/100* (n+1) ---> here n is the number of datas are present 
Q3= 75/100* (n+1) ---> same goes here 


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.boxplot(data["Purchase"])
```

here orange line is considered to be as median 
top line of square box is first quantile and bottom is third quantile 
whereas line is higher fence and lower line is lower fence --> here values which lies above the range will be consider as outliers and that has to be discarded


Inner Quantile range = Q3- Q1 
lower fence = Q1 - 1.5(inner quantile range )
higher fence = Q3 + 1.5 (inner quantile range)

any values which lies between the lower and higher fence will not be consider as outliers----> above the range will be consider as outliers 
that outliers will be removed/discard that values from the datasets 

```python
Q1=data["Purchase"].quantile(0.25)
Q3=data["Purchase"].quantile(0.75)
IQR=Q3-Q1
Q1,Q3,IQR
```

```python
LL = Q1 - 1.5*(IQR) 
UL = Q3 + 1.5*(IQR)
LL,UL
```

the values or the data which lies between -3431.0 and 21353.0 will be consider to be as outliers and it will be discarded

```python
#removing the otliers which lies between the range as mention above 
#Formula to remove the outliers 
data=data[(data["Purchase"]>=LL)& (data["Purchase"]<=UL)]
data
```
